```python
class Dict():
    def __init__(self):
        self.capacity = 6
        self.hash_table = [(None,None) for i in range(self.capacity)]
        
    def hash1(self,x):
        return x % self.capacity
        
    def put(self, key, value):
        hash_k = self.hash1(key)
        
        while self.hash_table[hash_k][0] != None:
            if self.hash_table[hash_k][0] == key:
                self.hash_table[hash_k] = (key,value)
                return
            else:
                hash_k = (hash_k + 1) % self.capacity
        self.hash_table[hash_k] = (key,value)
        
    def get(self, key):
        hash_k = self.hash1(key)
        h_key, value = self.hash_table[hash_k]
        if h_key == key:
            return value
        else:
            for i in range(self.capacity):
                hash_k = self.hash1(key+i)
                h_key, value = self.hash_table[hash_k]
                if h_key == key:
                    return value
        return None
    
class OuterNode():
    def __init__(self, key):
        self.key = key
        self.table = Dict()
        self.next = None
        
    def add(self,outernode):
        self.next = outernode
        
    def get(self, key2):
        return self.table.get(key2)
        
    def put(self, key2):
        current = self.table.get(key2)
        if current != None:
            self.table.put(key2, current + 1)
        else:
            self.table.put(key2, 1)
            
class OuterTable:
    M = 3
    ht = [None] * M
    
    def hash1(self, key):
        return key % self.M
    
    def get(self, key1, key2):
        i = self.hash1(key1)
        x = self.ht[i]
        while (x!=None):
            if (x.key==key1):
                return x.get(key2)
            else:
                x = x.next
        return None
    
    def put(self, key1, key2):
        i = self.hash1(key1)
        x = self.ht[i]
        while (x!=None):
            if (x.key==key1): 
                x.put(key2)
                return
            else:
                x = x.next
        n = OuterNode(key1)
        n.put(key2)
        n.add(self.ht[i])
        self.ht[i] = n 
```
